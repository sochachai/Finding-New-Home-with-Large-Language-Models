Imputation

```python
for column_name in ['crime_per_1000_residents',\
                    'assault_crime_rate','robbery_crime_rate','rape_crime_rate',\
                    'murder_crime_rate','total_violent_crime_rate','theft_crime_rate',\
                    'vehicle_theft_crime_rate','vehicle_theft_crime_rate','burglary_crime_rate',\
                    'arson_crime_rate','total_property_crime_rate','kidnapping_crime_rate',\
                    'drug_crime_rate','vandalism_crime_rate','identity_theft_crime_rate']:
    data[column_name] = data[column_name].fillna(np.mean(data[column_name]))
```

```python
data = data.drop(columns = ['vehicle_theft_crime_rate.1'])
```

Numeric normalization

```python
for column_name in ['crime_per_1000_residents',\
                    'assault_crime_rate','robbery_crime_rate','rape_crime_rate',\
                    'murder_crime_rate','total_violent_crime_rate','theft_crime_rate',\
                    'vehicle_theft_crime_rate','vehicle_theft_crime_rate','burglary_crime_rate',\
                    'arson_crime_rate','total_property_crime_rate','kidnapping_crime_rate',\
                    'drug_crime_rate','vandalism_crime_rate','identity_theft_crime_rate']:
    data[column_name] = data[column_name]/np.abs(np.max(data[column_name]))
```

```python
text_similarity_weight = 10
crime_per_1000_residents_weight = -0.5
assault_crime_rate_weight = -0.6
robbery_crime_rate_weight = -0.6
rape_crime_rate_weight = -0.8
murder_crime_rate_weight = -1.0
kidnapping_crime_rate_weight = -0.8
data['property_score'] = data['text_similarity']*text_similarity_weight + data['crime_per_1000_residents']*crime_per_1000_residents_weight\
                         + data['assault_crime_rate']*assault_crime_rate_weight + data['robbery_crime_rate']*robbery_crime_rate_weight\
                         + data['rape_crime_rate']*rape_crime_rate_weight + data['murder_crime_rate']*murder_crime_rate_weight\
                         + data['kidnapping_crime_rate']*kidnapping_crime_rate_weight
```

```python
#similarities.argsort()[0][::-1][:num_recommendations]
top_indices = data['property_score'].argsort()[::-1][0:5]
recommended_homes = data.iloc[top_indices]['link']
```

##Following is the recommendation using only text similarities

Text embeddings

```python
# Get embeddings
sentence_embeddings = bert.encode(data['description'].tolist())
# Define path to save embeddings
embeddings_path = "/content/drive/MyDrive/Recommendation_System/home_description_embeddings.npy"
# Save
np.save(embeddings_path, sentence_embeddings)
```

Establish recommendation

```python
sentence_embeddings = np.load(embeddings_path)
def home_recommendation(user_input,sentence_embeddings):
    print("Recommendations for \"", user_input, "\" are:", sep='')
    start = time.time()
    user_input_embedding = bert.encode([user_input])
    similarities = cosine_similarity(user_input_embedding,sentence_embeddings)
    print(similarities)
    num_recommendations = 5
    top_indices = similarities.argsort()[0][::-1][:num_recommendations]
    recommended_homes = data.iloc[top_indices]['link']
    print(recommended_homes)
    end = time.time()
    print(f"Time spent is {end-start}.")

input_list = ["shiny large garden spacious living room, 3 bathrooms, near school", "hardwood flooring new carpets remodelled kitchen",\
              "3 bedrooms 2 baths large garden 2 parking spaces bright easy commute"]
for i in input_list[2:3]:
    home_recommendation(i, sentence_embeddings)
    print()
```
